```python
# Support vector machine
from sklearn import svm
from sklearn.svm import SVR

# regr_svm = grid_svm.best_estimator_  # Best estimator based on hyperparameter tuning result

#=========== previously tuned 
model=regr_svm = SVR(C=50, 
               epsilon=0.00001, 
               gamma='auto',
               kernel='rbf').fit(xtrain, ytrain.values.squeeze())
#=---------------------------
m_used = 'SVR'

# predictions for train and test datasets
y_pred_test_1 = model.predict(xtest).flatten() # prediction for test dataset
y_pred_train_1 = model.predict(xtrain).flatten() # prediction for train dataset
# Predictions for all dataset and observed values
y_pred_all_1= model.predict(X).flatten() # predictions for all dataset (Remember X is the input)

# Invert
yy = dataset.iloc[:,-1:].values # observed values
y_pred_all_ = yy.min()+(yy.max()-yy.min()) * y_pred_all_1 # inverted predicted values
y_pred_test_ = yy.min()+(yy.max()-yy.min()) * y_pred_test_1 
y_pred_train_ = yy.min()+(yy.max()-yy.min()) * y_pred_train_1
y_observed = yy.min()+(yy.max()-yy.min()) * Y
y_observed_train = yy.min()+(yy.max()-yy.min()) *ytrain
y_observed_test = yy.min()+(yy.max()-yy.min()) *ytest

# Mean and standard deviation of predicted to observed outputs ratio
ratio_ = np.divide(y_pred_all_.reshape(y_pred_all_.shape[0], 1),  yy)
mean_ratio_ = round(np.mean(ratio_),3)
STD_ratio_ = round(np.std(ratio_),3)

print("Mean of predicted/experimental ratio is: " + str(mean_ratio_))
print("STD of predicted/experimental ratio is: " + str(STD_ratio_))

# R-square
from sklearn.metrics import r2_score
R2_full_dataset_  = r2_score(Y, y_pred_all_1)
R2_train_dataset_  = r2_score(ytrain, y_pred_train_1)
R2_test_dataset_  = r2_score(ytest, y_pred_test_1)

print("R-square for train dataset is: " + str(round(R2_train_dataset_,2)))
print("R-square for test dataset is: " + str(round(R2_test_dataset_,2)))
print("R-square for the whole database is: " + str(round(R2_full_dataset_, 2)))


# PLOT: OBSERVED VERSUS PREDICTED VALUES

# axis boundary for all plots:
lm1 = y_observed.values.max()
lm2 = y_pred_all_.max()
lm = 1.1 * max(lm1, lm2) # axis boundary
# I multiplied by 1.1 just to extend the axis limit from the maximum values

# plot on same figure
plt.figure(figsize=(6,6))  # plot size

# plt.scatter(y_observed, y_pred_all_, c='b', marker="o")
plt.scatter(y_observed_train.values, y_pred_train_, c='b', marker="o", label = 'train dataset')
plt.scatter(y_observed_test.values.flatten(), y_pred_test_, c='r', marker="o", label = 'test dataset')
# plt.legend()
plt.ylabel('Predicted value for PHL', fontsize = 14)
plt.xlabel('Experimental value for PHL', fontsize = 14)

# change axis limit
axes = plt.gca()
axes.set_xlim([0, lm])
axes.set_ylim([0, lm])
# peect match line
x = [0, lm]
y = x
plt.plot(x, y, 'r')

# safe prediction lines
plt.plot(x, np.multiply(y, 1.2), 'g')
plt.plot(x, np.multiply(y, 0.8), 'g')

# add text: Display R-square in the graph:
plt.text(0.3*lm, 0.7*lm, r'$R^2$ = ' + str(round(R2_full_dataset_,3)), horizontalalignment='center',
     verticalalignment='center', fontsize=14)

plt.legend(loc="best")


# save the figure to a file
plt.savefig('Results/Figures/'+m_used+'.tiff',format='tiff', dpi=1000,bbox_inches = "tight")
plt.show();
```
